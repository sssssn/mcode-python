## 2) 이상치 확인 및 처리

## 상자그림 활용 (이상치:Q1, Q3로부터 1.5*IQR 초과하는 값)

```python
# 타이타닉 데이터 불러오기
df = sns.load_dataset('titanic')

# (참고) 상자그림
sns.boxplot(df['age'])
```

```python
# Q1, Q2, Q3 구하기
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

```python
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
print(upper, lower)
```

```python
# 문제 : age 변수의 이상치를 제외한 데이터 수는? (상자그림 기준)
cond1 = (df['age'] <= upper) # 부호 주의
cond2 = (df['age'] >= lower) # 부호 주의
print(len(df[cond1 & cond2]))
print(len(df[cond1]))
print(len(df))
```

```python
# 문제 : age 변수의 이상치를 제외한 데이터셋 확인 (상자그림 기준)
df_new = df[cond1 & cond2]
df_new
```

## 표준정규분포 활용 (이상치: ±3Z 값을 넘어가는 값)

```python
# 데이터 표준화, Z = (개별값 - 평균) / 표준편차
```

```python
mean_age = df['age'].mean()
std_age = df['age'].std()
print(mean_age)
print(std_age)
```

```python
znorm = (df['age'] - mean_age) / std_age
znorm
```

```python
# 문제 : 이상치의 개수는 몇 개인가? (±3Z 기준)
```

```python
cond1 = (znorm > 3)
len(df[cond1])
```

```python
cond2 = (znorm < -3)
len(df[cond2])
```

```python
print(len(df[cond1]) + len(df[cond2]))
```

# ✅ 7. 날짜/시간 데이터, index 다루기

## 1) 날짜 다루기

```python
# 데이터 만들기
df = pd.DataFrame( {
    '날짜' : ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230511'],
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A'],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600] })
df
```

```python
df.info()
```

```python
# 데이터 타입 datetime 으로 변경
df['날짜'] = pd.to_datetime(df['날짜'])
df.info()
```

```python
# 년, 월, 일 변수(열) 추가하기
df['year'] = df['날짜'].dt.year
df['month'] = df['날짜'].dt.month
df['day'] = df['날짜'].dt.day
df
```

```python
# 날짜 구간 필터링
df [df['날짜'].between('2023-01-01', '2023-02-23')] # 좌우 모두 포함
# (주의) 날짜와 시간이 같이 있는 데이터에 between 함수를 사용할 경우, 동일한 형식으로 사용
# (예) 2023-01-05 12:30:05 => between('2023-01-05 12:00:00', '2023-01-05 12:40:00') OK
#                          => between('2023-01-05', '2023-01-05') NG
```

```python
# 날짜를 인덱스로 설정 후 loc 함수 사용
# 데이터 만들기
df = pd.DataFrame( {
    '날짜' : ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230511'],
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A'],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600] })

# 데이터 타입 datetime 으로 변경 (필수)
df['날짜'] = pd.to_datetime(df['날짜'])

df = df.set_index('날짜')
# df = df.set_index('날짜', drop = False) drop=True(기본) or False
df.head(3)
```

```python
print(df.loc['2023-01-05' : '2023-02-23']) # 둘 다 기간 포함
print(df.loc[(df.index >= '2023-01-05') & (df.index <= '2023-02-23')])
```

## 2) 시간 다루기

```python
# 시간 데이터 만들기
df = pd.DataFrame( {
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A'],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600] })
time = pd.date_range('2023-09-24 12:25:00', '2023-09-25 14:45:30', periods = 7)
df['time'] = time
df = df[['time', '물품', '판매수', '개당수익']]
df
```

```python
# 1. 시간이 '변수'로 있을 경우 (between 함수)
df = df [df['time'].between('2023-09-24 12:25:00', '2023-09-24 21:11:50')]
# (주의) 날짜와 시간이 같이 있는 데이터에 between 함수를 사용할 경우, 동일한 형식으로 사용
# (예) : 2023-01-05 12:30:05 => between('2023-01-05 12:00:00', '2023-01-05 12:40:00') OK
#                          => between('2023-01-05', '2023-01-05') NG

print(df)
```

```python
# 2. 시간이 'index'로 있을 경우 (between_time, loc 함수)

# 1) df.between_time(start_time = '00:00:00', end_time = '00:00:00')
# ★ 날짜와 상관없이 특정 시간대를 필터링 해야 할 때
# (★주의: 시간이 index에 위치해야 함)

# index 새로 지정
df = df.set_index('time')
# print(df)

df = df.between_time(start_time = '12:25:00', end_time = '21:11:50') # 시작, 끝 시간 모두 포함
# include_start = False, include_end = False 옵션으로 시작, 끝 시간 제외 가능
print(df)

# (참고사항)
# index 초기화 (인덱스를 칼럼으로)
# df = df.reset_index()
```

```python
# 2) loc 함수 사용
# (★주의: 시간이 index에 위치해야 함)

print(df.loc['2023-09-24 12:25:00' : '2023-09-24 21:11:50']) # 둘 다 포함
print(df.loc[(df.index >= '2023-09-24 12:25:00') & (df.index <= '2023-09-24 21:11:50')])
```
